```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.get_dataset_names()
```

```python
#Loading the Data
mpg = sns.load_dataset('mpg')
```

```python
#Data Exploration
mpg.shape
mpg.describe
```

```python
mpg
```

```python
#Correlation of variables

mpg.corr(numeric_only=True) #because there are characters in our dataset
```

```python
sns.heatmap(mpg.corr(numeric_only=True)) #the lighter, the stronger the positive relationship
#the darker the stronger the negative correlation
```

```python
#Scatter Plot
sns.set_theme()
sns.set_style('white') #'darkgrid' 'whitegrid' 'ticks' 'dark' 'white'
sns.set_context('paper', font_scale=1.2) #'paper' 'poster' 'talk'
plt.figure(figsize=(10,5)) #size of the figure
sns.scatterplot(data=mpg, x='weight', y='mpg', hue='origin', palette='mako')
sns.despine(left=False, right=True) #removes the gridlines around the rectangule
```

```python
#color palettes
# https://seaborn.pydata.org/tutorial/color_palettes.html
#https://matplotlib.org/stable/gallery/color/named_colors.html
```

```python
#Line Plot

sns.lineplot(data=mpg,x='model_year',y='mpg')
```

```python
#Point Plot

sns.pointplot(data=mpg,x='model_year',y='mpg')
```

```python
#KDE Plot

sns.kdeplot(data=mpg, x='weight', y='mpg')
```

```python
sns.kdeplot(data=mpg, x='weight', y='mpg', fill=True)
```

```python
#The heavier the vehicle more power it needs to accelerate thus impacting the miles per gallon of the car. Since increase in the weight of the car decrease the miles per gallon of the car, hence the correlation between these two variables is negative.
```
